```python
import json

d = ""
with open('camout.json') as json_data:
    d = json.load(json_data)
    print(d)
```

```python
d = sorted(d, key=lambda k: k['size'], reverse=True)
```

```python
tmp = d
index = 0
size = 0
data = []
for song in tmp:
        if index < 20:
            print (song['size'])
            index = index + 1
            size = int(song['size'])
            data.append(song)
#            data[0]['size'] = song['size']
            
            
```

```python
print (data)
```

```python
with open('camout_top.json', 'w') as outfile:
    json.dump(data, outfile)
```
